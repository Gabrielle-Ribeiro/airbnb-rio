**Lidando com Dados Nulos**

Verificando quais colunas tem dados nulos.

```python
df.isna().sum()
```

*Coluna reviews_per_months*

A coluna *reviews_per_month* (avaliações por mês) tem 8252 linhas nulas. Isso representa aproximadamente 35% da quantidade total de dados. É uma quantidade muito expressiva para descartar todas essas linhas.

Vamos assumir que como não há uma quantidade de avaliações por mês, então não foi feita nenhuma avaliação para aquele Airbnb. O que significa que podemos preencher essas linhas com o valor 0.

```python
df['reviews_per_month'] = df['reviews_per_month'].fillna(0)
```

*Coluna bathrooms_text*

A coluna *bathrooms_text* (textos sobre banheiros) tem 68 linhas nulas. Como isso representa menos de 1% do total de dados, vamos só excluir as linhas que não possuem informação sobre banheiro.

```python
df = df.dropna(subset=['bathrooms_text'], axis=0)
```

*Coluna bedrooms*

A coluna *bedrooms* (quartos) possui 1488 linhas nulas. Isso representa aproximadamente 6% dos dados. Se verificarmos os valores que essa coluna tem, vemos que a menor quantidade de quartos existente é 1. Então vamos preencher esses dados com esse valor, já que aparentemente todo airbnb possui pelo menos 1 quarto.

```python
len(df[df['bedrooms'].isnull() & df['beds'].isnull()])
```
